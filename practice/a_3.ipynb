```python
# loading the required dataset.
df2 = pd.read_csv('../data/loan.csv')
df2
```

```python
# getting the statistical description.
df2.describe()
```

```python
# getting the information about the dataset.
df2.info()
```

```python
# now determining the columnwise null values count.
df2.isnull().sum()
```

```python
# dealing with the null values and converting the datatypes.
df2.dropna(axis=0, how='any', inplace=True)
df2['Total_Income'] = df2['Total_Income'].str[1:]
df2['Total_Income'] = df2['Total_Income'].str.replace(',', '')
df2['Total_Income'] = df2['Total_Income'].astype('float')
print('The dataframe after removing the null field rows is')
df2
```

```python
# now re-checking the df2 if null values exists.
df2.isnull().sum()
```

```python
# now grouping the data for gender and displaying the statistical description.
grouped = df2.groupby('Gender')
for group, data in grouped:
    print('-_-_-_-_-_-_-_-_-_-_-_-_-_-_-_-_-_-_-_')
    print(f'For {group}')
    print(f'Statistical Description:')
    print(data.describe())
```

```python
# statistical description for numerical variables grouped by the categorical variables.
categorical_columns = df2.select_dtypes(include=['object']).drop(columns=['Loan_ID']).columns
numerical_columns = df2.select_dtypes(include=['number']).columns
summary_stats = df2.groupby(categorical_columns.tolist())[numerical_columns].describe()
summary_stats.head(20)
```
