```python
# importing the dependencies.
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder
```

```python
# loading the dataset.
na_values = ['NaN', 'N.A', 'na']
df = pd.read_csv('../data/academic_data.csv', na_values=na_values)
df
```

```python
# describing the statistical description.
df.describe()
```

```python
# information about the dataset.
df.info()
```

```python
# now checking the datatypes of all the columns.
df.dtypes
```

```python
# checking the dimensions and the size of the dataset.
print(f'The no.of rows are: {df.shape[0]} \nThe no.of columns: {df.shape[1]}')
print(f'The total no.of elements are: {df.size}')
```

```python
# now checking the null values.
df.isnull().sum()
```

```python
# dealing with the null values.
df['DSBDA'] = df['DSBDA'].fillna(df['DSBDA'].mean())
df['SPOS'] = df['SPOS'].fillna(df['SPOS'].mean())
df['WT'] = df['WT'].fillna(df['WT'].mean())
df['DA'] = df['DA'].fillna(df['DA'].mean())
print(f'The null count after dealing is: {df.isnull().sum()}')
```

```python
# converting categorical variables into quantitative.
le = LabelEncoder()
df['Gender'] = le.fit_transform(df['Gender'])
print('Dataframe after encoding')
df
```
