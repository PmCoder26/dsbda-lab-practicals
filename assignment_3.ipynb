```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('data/loan.csv')
df
```

```python
# Statistical summary.
df.describe()
```

```python
df.isnull().sum()
```

```python
df['Dependents'].fillna(0, inplace=True)
df['Self_Employed'].fillna('No', inplace=True)
df['LoanAmount'].fillna(0, inplace=True)
df['Loan_Amount_Term'].fillna(0, inplace=True)
df['Credit_History'].fillna(0, inplace=True)
df.dropna(axis=0, inplace=True)
df.isnull().sum()
```

```python
df['Gender'] = df['Gender'].map({'Male': 1, 'Female': 0})
df['Married'] = df['Married'].map({'Yes': 1, 'No': 0})
df['Education'] = df['Education'].map({'Graduate': 1, 'Not Graduate': 0})
df['Loan_Status'] = df['Loan_Status'].map({'Y': 1, 'N': 0})
df['Self_Employed'] = df['Self_Employed'].map({'Yes': 1, 'No': 0})
df['Property_Area'] = df['Property_Area'].map({'Urban': 0, 'Rural': 1, 'Semiurban': 2})
df
```

```python
genderwise_max_total_income = df.groupby('Gender')['Total_Income'].max()
genderwise_min_totoal_income = df.groupby('Gender')['Total_Income'].min()
print("Genderwise max total income", genderwise_max_total_income)
print("Genderwise min total income", genderwise_min_totoal_income)
```
